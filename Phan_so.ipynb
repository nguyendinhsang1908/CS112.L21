# 5.Programing

```python
import math
a=int(input())
b=int(input())
c=int(input())
d=int(input())
index=0
while a/b!=c/d:
  if(a/b>c/d):
    index=0
    break
  else:
    a+=1
    b+=1
    ucln=math.gcd(a,b)
    a=a//ucln
    b=b//ucln
    index+=1
print(index)
```
